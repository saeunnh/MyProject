# Thermoelastic equations 

Let $\Omega$ be a regular domain with boundary $\partial \Omega$. We indicate the outward unit normal vector of $\partial \Omega$ by $\mathbf{n}$.<br>
The equations formulation read:
$$ \nabla \cdot \frac{1}{2}D(\nabla \mathbf{u} + \nabla \mathbf{u}^T) - \beta _S K  \nabla (T_m-T_0)\mathbf{I} = F \\
\rho_s C_s  \frac{\partial T_m}{\partial t} - \nabla \cdot (\kappa_S \nabla T_m) - \beta _S K T_m \frac{\partial (\nabla \cdot \mathbf{u})}{\partial t}\mathbf{I} = f$$

Linerized version of the equation can be written:
$$ \frac{\partial}{\partial t} [\rho_s C_s T_m  - \beta _S K T'_m (\nabla \cdot \mathbf{u}) ] - \nabla \cdot (\kappa_S \nabla T_m) = f $$

Which is identical to the simplest version of the Biot problem and can be solved with

Dirichlet boundary conditions $\partial \Omega_d$:
$$ T_m = T_b \quad  \quad , \quad \quad \mathbf{u} = \mathbf{u}_b $$

and Neumann boundary conditions $\partial \Omega_n$:
$$ - \kappa_S \nabla (T_m-T_0) \cdot \mathbf{n} = h_b \quad ,\quad \frac{1}{2}D(\nabla \mathbf{u} + \nabla \mathbf{u}^T) \cdot \mathbf{n} = t_b$$


In the above, the unknowns are the displacements $\mathbf{u}$ and the temperature $T$. The parameters are the scalar and vector source/sink terms $F$ and $f$. 
Where $\kappa$ is the average thermal conductivity tensor of the rock matrix, $D$ is the stiffness matrix, the volumetric thermal expansion coefficient $\beta$, the bulk moduli $K$, the specific heat capacity $C$ and density $\rho$ of the rock matrix. <br>

$T_b$ is the temperature at the boundary (Dirichlet condition), $h_b$ is the heat flux at the boundary 
(Neumann condition), $\mathbf{u}_b$ is the displacement at the boundary (Dirichlet condition) 
and $t_b$ is the traction at the boundary (Neumann condition).

Methods to assign data and initial conditions. We will mainly use default values for parameters, while overriding some of the values. 

```python
def assign_data(g, d, parameter_keyword_flow, parameter_keyword_mechanics):
    
    # Assign a non-default permeability and thermal conductivity, for illustrative purposes
    k = 2 * np.ones(g.num_cells)   
    kappa = pp.SecondOrderTensor(k)
    # fluid_compressibility = 0.1 * np.ones(g.num_cells)
    
    
    # Define boundary type: Dirichlet at top and bottom t=[T_0,T_M], Neunam at right and left flux=0
#    bound_faces = g.get_all_boundary_faces()
#    print(bound_faces) #     [ 0  4  5  9 10 14 15 19 20 21 22 23 36 37 38 39]
#    top_faces = np.ravel(np.argwhere(g.face_centers[1] > Ly - 1e-10))
#    top_faces = [36 37 38 39]
#    bot_faces = np.ravel(np.argwhere(g.face_centers[1] < 1e-10))
#    bot_faces = [20 21 22 23]
#    lef_faces = np.ravel(np.argwhere(g.face_centers[0] < 1e-10))
#    lef_faces = [0 5 10 15]
#    rig_faces = np.ravel(np.argwhere(g.face_centers[0] > Lx - 1e-10))
#    rig_faces = [4 9 14 19]
    
    
    neuman = np.block([lef_faces,rig_faces])
    dirich = np.block([top_faces,bot_faces])

    bound_faces = np.block([neuman,dirich])
    labels = np.array(['dir']*bound_faces.size)
    labels[0:len(neuman)] = 'neu' 
    t_bound = np.zeros(g.num_faces)
    t_bound[top_faces] = T_0 
    t_bound[bot_faces] = T_M 

    
    
#    bound_faces = g.get_all_boundary_faces()
#    bound_flow = pp.BoundaryCondition(g, bound_faces.ravel("F"), ["dir"] * bound_faces.size)
    bound_flow = pp.BoundaryCondition(g, bound_faces, labels)    
    
    
    bound_mech = pp.BoundaryConditionVectorial(
        g, bound_faces.ravel("F"), ["dir"] * bound_faces.size
    )
#    boundary_values_flow = np.zeros(g.num_faces)
#    boundary_values_flow[0] = 1
    
    
    # Create a dictionary to override the default parameters
    t = T_0 #(ætti e.t.v. að vera  np.ones(g.num_cells))
    alfa = 1
    alfa_heat_flow = alfa*t

    # Fyrir ofan er bæði bound_flow og boundary_values_flow sem notað er í bc bc_values 
    specified_parameters_flow = {"second_order_tensor": kappa, "biot_alpha": alfa_heat_flow,
                                 "bc": bound_flow, "bc_values": t_bound,
                                "time_step": 0.3, "mass_weight": 1} #fluid_compressibility}
    specified_parameters_mechanics = {"biot_alpha": alfa, "bc": bound_mech}
    
    # Initialize the parameters for flow and transport.
    # The following keywords are used to define which set of default parameters to pick, i.e. the
    # type of parameters. For example, initialize_default_data initializes a second order permeability
    # tensor in the flow data, and a fourth order stiffness tensor in the mechanics data.
    parameter_type_flow = "flow"
    parameter_type_mechanics = "mechanics"
    pp.initialize_default_data(g, d, parameter_type_flow, specified_parameters_flow,
                               parameter_keyword_flow)
    pp.initialize_default_data(g, d, parameter_type_mechanics, specified_parameters_mechanics,
                               parameter_keyword_mechanics)
    pp.set_state(d, {parameter_keyword_mechanics: {'bc_values': np.zeros(g.dim * g.num_faces)}})
    return bound_flow, t_bound
```

```python
def initial_condition(g, d, kw_m, kw_f, variable_m, variable_f):
    d_0 = np.zeros((g.dim, g.num_cells))
    d_0[0,:] = 0
    d_0 = d_0.ravel('F')
    t_0 = T_M * np.ones(g.num_cells)
    #p_0[-1] = 2
    state = {variable_m: d_0, variable_f: t_0}
    pp.set_state(d, state)
    return d_0, t_0
```

## Problem setup
The practical way of setting up a problem on a single grid with two variables and multiple terms is described here. First, we define a grid and assign data, which we tag with a keyword. The purpose of this keyword is for the discretization operators (such as `pp.Mpsa()` below) to identify which parameters to use. For more details and explanations, see the tutorial on parameter assignment.

```python
Nx = Ny = 4
Lx = Ly = 40
phys_dims = [Lx,Ly]
#g = pp.CartGrid([Nx, Ny], phys_dims)
def make_grid(nn=[Nx, Ny]):
    # Define a grid for a toy problem
    gb = pp.meshing.cart_grid([], nn)
    gb.compute_geometry()
    gb.assign_node_ordering()
    return gb
```

```python
# Define a grid
gb = make_grid()
g = gb.grids_of_dimension(2)[0]
data = gb.node_props(g)
print(data)

cell_id = np.arange(g.num_cells)

pp.plot_grid(g, cell_value=cell_id, info='cf', alpha=0.0, figsize=(15,12))
```

```python
# Define a grid
gb = make_grid()
g = gb.grids_of_dimension(2)[0]
data = gb.node_props(g)
# Assign data, tagging them with the following keywords
kw_f = "flow"
kw_m = "mechanics"
variable_f = "pressure"
variable_m = "displacement"
#print(bound_flow)
#print(boundary_values_flow)
bound_flow, boundary_values_flow = assign_data(g, data, kw_f, kw_m)
print(bound_flow)
print(boundary_values_flow)

d_0, t_0 = initial_condition(g, data, kw_m, kw_f, variable_m, variable_f)
print('Initial conditions')
print(d_0)
print(t_0)
data[pp.STATE][variable_m] = d_0
data[pp.STATE][variable_f] = t_0
```

Now, discretize the mechanics related terms using the Biot class. This is a dirty trick to ensure efficient discretization where we re-use the local gradient systems of MPSA (see pp/numerics/fv/mpsa.py) for all the mechanical terms involved.

```python
biot_discretizer = pp.Biot(kw_m, kw_f, variable_m, variable_f)
biot_discretizer._discretize_mech(g, data)
```

## Assembly
The equations we assemble are a *discretized* version of the Biot equations. This requires a spatial discretization of the individual terms, taken care of by objects such as e.g. `pp.Mpsa` for the $\nabla \cdot \frac{1}{2}D(\nabla \mathbf{u} + \nabla \mathbf{u}^T)$ term, see below. However, a *temporal* discretization must also be chosen, with some implications on the assembly. In this case, we use the Implicit Euler, and therefore need to change the standard PorePy discretization objects for the mass term $\frac{\partial}{\partial t}\beta p$ and the flow term $\nabla \cdot K \nabla p$, so that hte former returns a term to the right hand side, whereas the latter must be multiplied by the time step. Classes that make this modifications (essentially wrappers around standard discretizations) are available by the imported IE_disrcetization.

The assembly requires some information about the relationship between variables, terms (of the equation) and discretization objects to be stored in the data dictionary `d`. The information includes the names of the primary variables and identification of the terms of the equations and which objects to discretize them with. Note that the numbering of the terms (i and j in "term_ij") corresponds to the Biot equation as defined above, since stress equals stiffness tensor times the symmetric gradient of displacement.

```python
# Variable names
v_0 = variable_m
v_1 = variable_f
# Names of the five terms of the equation + additional stabilization term.
#                                        Term in the Biot equation:
term_00 = "stress_divergence"          # div symmetric grad u
term_01 = "pressure_gradient"          # alpha grad p
term_10 = "displacement_divergence"    # d/dt alpha div u
term_11_0 = "fluid_mass"               # d/dt beta p
term_11_1 = "fluid_flux"               # div (rho g - K grad p)
term_11_2 = "stabilization"            #

# Store in the data dictionary d and specify discretization objects.
data[pp.PRIMARY_VARIABLES] = {v_0: {"cells": g.dim}, v_1: {"cells": 1}}
data[pp.DISCRETIZATION] = {
    v_0: {term_00: pp.Mpsa(kw_m)},
    v_1: {
        term_11_0: IE_discretizations.ImplicitMassMatrix(kw_f, variable_f),
        term_11_1: IE_discretizations.ImplicitMpfa(kw_f),
        term_11_2: pp.BiotStabilization(kw_f, v_1),
    },
    v_0 + "_" + v_1: {term_01: pp.GradP(kw_m)},
    v_1 + "_" + v_0: {term_10: pp.DivU(kw_m, kw_f, v_0)},
}
```

Assembly of the global linear system can in many cases be carried out by a single function call of the assembler. Note that for the current problem, this is not possible. Rather, we had to discretize first by calling `biot_discretizer._discretize_mech(g, d)`.
Below, A is the global linear system and b is the corresponding right hand side. We obtain the displacement and pressure solution by solving the linear system. We solve the transient system for three time steps, updating the solution vectors in the data dictionary for each step, and also storing all solutions for examination.

```python
assembler = pp.Assembler(gb)
# Discretize the flow and accumulation terms - the other are already handled by the biot_discretizer
assembler.discretize(term_filter=[term_11_0, term_11_1])

time_steps = np.arange(3)
stored_temperature = []
stored_displacements = []
for _ in time_steps:
    A, b = assembler.assemble_matrix_rhs()
    x = sps.linalg.spsolve(A, b)
    # Distribute the variables (pressure and displacement) on the data
    # dictionaries of the grid bucket
    assembler.distribute_variable(x)
    displacement = data[pp.STATE][variable_m]
    pressure = data[pp.STATE][variable_f]
    pp.set_state(data, {variable_m: displacement, variable_f: pressure})
    stored_temperature.append(pressure)
    stored_displacements.append(displacement)
```

## Visualization
We visualize the initial condition and the final solution cell-wise using PorePy's plotting functionality, representing the pressure by colors and displacement by arrows. To plot the displacement vectors, we first need to reshape them and add a zero value for the z direction. We also scale the values using the `vector_scale` argument for visualization purposes.

```python
d_0 = np.reshape(d_0, (2, g.num_cells), 'F')
d_0 = np.vstack((d_0, np.zeros(g.num_cells)))
pp.plot_grid(g, t_0, d_0, figsize=(8, 6), vector_scale=1/3, if_plot=False)
plt.title("Initial condition")
for t in time_steps:
    d = np.reshape(stored_displacements[t], (2, g.num_cells), 'F')
    d = np.vstack((d, np.zeros(g.num_cells)))
    pp.plot_grid(g, stored_temperature[t], d, figsize=(8, 6), vector_scale=1/2, if_plot=False)
    plt.title("Time step {}".format(str(t + 1)))
plt.show()
```
